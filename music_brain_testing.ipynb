```python
#Set the username and password to be used in subsequent queries to the MusicBrainz XML API that require authentication(not sure if this is necessary if I'm only downloadin)

# mb.auth('username','password')
```

```python
#Sets the rate limiting behavior of the module. Must be invoked before the first Web service call. If the limit_or_interval parameter is set to False then rate limiting will be disabled. If it is a number then only a set number of requests (new_requests) will be made per given interval (limit_or_interval).
mb.set_rate_limit(limit_or_interval=1,new_requests=5)
```

```python
#Set the User-Agent to be used for requests to the MusicBrainz webservice. This must be set before requests are made.
mb.set_useragent('Chrome','86.0.4240.111')

# Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36
```

```python
# Sets the format that should be returned by the Web Service to json
mb.set_format(fmt='json')
```

```python
#artist search with music brainz
artist = "Ricky Martain"

results=mb.search_artists(artist)

results

```
